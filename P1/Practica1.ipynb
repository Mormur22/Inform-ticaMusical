```python
import numpy as np
import matplotlib.pyplot as plt

sampling_rate = 1
duration = 1
time = np.arange(0, duration, 1/sampling_rate)
frequency = 1
wave = np.sin(2 * np.pi * frequency * time)

# Dibujar la señal
plt.plot(time, wave)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Sine Wave')
plt.show()
```

```python

# Frecuencia de muestreo
fs = 1

# Duración de la señal
duration = 1

# Número de muestras
n_samples = duration * fs

# Eje de tiempo
time = np.linspace(0, duration, n_samples, endpoint=False)

sample = np.sin(2 * np.pi * 10 * time)



def vol(sample, vol):
    return sample * vol

def modulaVol(sample, fs):
    time = np.arange(len(sample)) / len(sample)
    modulation = np.sin(2 * np.pi * fs * time)
    return sample * modulation

modulated_sample = modulaVol(sample, 5)


plt.plot(time, sample, label='Original signal')
plt.plot(time, modulated_sample, label='Modulated signal')
plt.legend()
plt.show()
```
